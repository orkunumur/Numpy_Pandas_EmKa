8) Oluşturmuş olduğunuz df_male ile çalışarak elimizdeki verideki erkeklerin yüzde kaçının sigara içtiğini bulun. Aynı şekilde elimizdeki kadınların yüzde kaçının sigara içtiğini de bulun. Aralarında anlamlı bir fark mı yorumlayın.

```python
smoker = df_male.value_counts("smoker")
smoker_Female = df_female.value_counts("smoker")
```

```python
smoke_yes = (smoker[0] / df_male.shape[0]) * 100 
smk_Fe_yes = (smoker_Female[0] / df_female.shape[0]) * 100
print("Sigara İçen Erkek Yüzdesi:", smoke_yes, "\n" "Sigara İçen Kadın Yüzdesi: ", smk_Fe_yes)
# Anlamlı bir fark yok.
```

```python
df_male.value_counts("smoker", normalize=True)
```
